## Importar librerías

```
import numpy as np
```

## Fijar semilla del generador de números aleatorios.

```
#seed(1234)
```

Se generan un conjunto de datos, primeramente las X y se hacen pasar por la función $y=2\cdot x -3$. Además se añade un poco de ruido 

```
X=np.arange(-10,10,0.1)
```

```
def function_NumberGenerator(X):
  #from  random import random
  y= 2*X-3.0
  return y
```

```
Y=function_NumberGenerator(X)
Y=Y+20.*np.random.random_sample(len(Y))
```

```
print(len(X),len(Y))
```

Se visualizan los datos generados.

```
plt.figure(1, figsize=(8,8))

plt.scatter(X, Y, color="black")
plt.grid()
plt.xlim(-10.1,10.1)
plt.ylim(-25.1,40.1)

plt.ylabel('Y')
plt.xlabel('X')
```

Se redimensionan el vector X por compatibilidad con el modelo.

```
print(X.shape)
X = X[:, np.newaxis]
Y = Y[:, np.newaxis]
print(X.shape)
```

## Se establece el modelo

```
regr = linear_model.LinearRegression()
```

Y se ajusta a los datos.

```
regr.fit(X, Y)
```

Se hace la predicción.

```
y_pred = regr.predict(X)
```

## Visualización de los resultados

```
plt.figure(1, figsize=(8,8))

plt.scatter(X, Y, color="black")
plt.scatter(X, y_pred, color="red")
plt.grid()
plt.xlim(-10.1,10.1)
plt.ylim(-25.1,40.1)

plt.ylabel('Y')
plt.xlabel('X')
```

Se imprimen los coeficientes del modelo.

```
print('Coefficients: \n', regr.coef_, regr.intercept_)
```

Se muestran los errores.

```
print('Error Cuadrático medio: %.2f'
      % mean_squared_error(Y, y_pred))
```

```
plt.figure(1, figsize=(8,8))

plt.hist(Y-y_pred, bins=20)
plt.xlim(-12,12)
```

## Propuestas adicionales.

*   Hacer un ejemplo de regresión lineal bivariante tanto en la entrada como en la salida.

```
regr = linear_model.LinearRegression()
regr.fit(X, Y)
y_pred = regr.predict(X)

plt.figure(1, figsize=(8,8))

plt.scatter(X, Y, color="black")
plt.scatter(X, y_pred, color="red")
plt.grid()
plt.xlim(-10.1,10.1)
plt.ylim(-25.1,40.1)

plt.ylabel('Y')
plt.xlabel('X')

####### x fitted over Y
regr2 = linear_model.LinearRegression()
regr2.fit(Y, X)
X_pred = regr2.predict(Y)

plt.scatter(X, Y, color="black")
plt.scatter(X_pred, Y, color="green")
plt.grid()
plt.xlim(-10.1,10.1)
plt.ylim(-25.1,40.1)

plt.ylabel('Y')
plt.xlabel('X')
plt.legend(["Fit Y(x)","Fit X(Y)"])
```
